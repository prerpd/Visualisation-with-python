# Introduction to Matplotlib

```python
#Putting Airtist layer to use

from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas #import FigureCanvas
from matplotlib.figure import Figure   #import Figure artist
fig = Figure()
canvas = FigureCanvas(fig)

#create 10000 random numbers using numpy
import numpy as np
p = np.random.randn(10000)

axis = fig.add_subplot(111) #creates an axix artist
axis.hist(p, 100)     #create histogram of generated numbers
axis.set_title('Normal distribution with $\mu = 0, \sigma=1$')
#saves the image to a png file
fig.savefig('matplotlib_hist.png')
```

```python
# Previous exercise with pypolt
#%matplotlib inline
#Magic functions starts with % sign. Here inline command is being used to generate plots inline 
#if the plots are opening in a separate window
#limitation of Backend - Image can not be modified once it is rendered
#%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.hist(p, 100)
plt.title('Normal distribution with $\mu = 0, \sigma=1$')
plt.show()
```

## Line plot exercise

```python
df_can.head()
```

```python
haiti = df_can.loc['Haiti', years]  #This command will extract the migrants yearwise
haiti.head()
```

```python
haiti.plot()   #to plot the year vs no of immigration
```

```python
# the year index in haiti is string type. Following command converts it into 
haiti.index = haiti.index.map(int)
#setting the kind of plot
haiti.plot(kind = 'line')

plt.title('Immigration from Haiti')
plt.xlabel('Years')
plt.ylabel('Immigrants number')
```

```python
#for annotating the spike in the plot
haiti.plot(kind= 'line')

plt.title('Immigration form Haiti')
plt.xlabel('Years')
plt.ylabel('Immigrant numbers')

#plt.text(x axis point, y axis point, 'text to annotate')
# In case x value is string, need to mention the x index value
#of the point where annotation is to be placed

plt.text(2000, 6000, '2010 Earthquake')
plt.show()
```

```python
df_CI = df_can.loc[['India', 'China'], years]
df_CI.head()
```

```python
# transpose function is used to transpose the data in the matrix.
#as can be seen below, the row and column has been inerchanges.
# Here df_CI is a data frame hence we need to transpose it.
# Earlier haiti was just a series hence we did not need to transpose it.
df_CI = df_CI.transpose()
df_CI.head()
```

```python
#plotting India and China immigration
df_CI.index = df_CI.index.map(int)
df_CI.plot(kind = 'line')
# other kinds that can be included in plot
#bar for vertical bar plots
#barh for horizontal bar plots
#hist for histogram
#box for boxplot
#kde or density for density plots
#area for area plots
#pie for pie plots
#scatter for scatter plots
#hexbin for hexbin plot

plt.title('Immigration from India and China')
plt.ylabel('Immigrant numbers')
plt.xlabel('Years')
```

```python
df_can.head()
```

```python
df_can.sort_values(by = 'Total', ascending = False, axis = 0, inplace = True)
# inplace makes the canges in the existing dataframe
df_top5 = df_can.head(5)
df_top5.head()
```

```python
df_top5 = df_top5[years].transpose()
df_top5.head()
```

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind = 'line', figsize= (14, 8))
#figsize set the size of the figure as x, y size values
plt.title('Immigration from top 5 countries')
plt.ylabel('Immigration number')
plt.xlabel('Years')
```

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind = 'box', figsize= (16, 8))
#figsize set the size of the figure as x, y size values
#plt.title('Immigration from top 5 countries')
#plt.ylabel('Immigration number')
#plt.xlabel('Years')
```
